#### Libraries Importation

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv(r"C:\Users\User\Documents\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df
```

#### Data Cleaning

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0).astype(float)
df
```

```python
df.describe()
```

#### Churn Percentage

```python
df_churn = df.groupby('Churn').size()
df_churn = df_churn.reset_index(name= 'count')
num = len(df)
df_churn['churn percentage'] = (df_churn['count']/num)*100
df_churn
```

```python
df_churnn = df[df['Churn'] == 'Yes']
df_churnn
```

```python
df_t1 = df[df['Churn'] == 'No']
df_t1
```

#### Churn percentage by Internet Service

```python
df_t2 = df.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount')
df_tnc = df_t1.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount_nochurn')
df_tc = df_churnn.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount_churn')
df_t3 = df_t2.merge(df_tnc, on= 'InternetService', how= 'left').merge(df_tc, on= 'InternetService', how= 'left')
df_t3['%churn_typewise'] = (df_t3['internetcount_churn']/df_t3['internetcount'])*100
df_total_churn = len(df_churnn)
df_t3['%churn_wrt_Gc'] = (df_t3['internetcount_churn']/df_total_churn)*100
df_t3

```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Internet Service Distribution & Churn Insights', fontsize=14, fontweight='bold')

# Global distribution
sns.barplot(data=df_t3, x='InternetService', y='internetcount', ax=axes[0], hue='InternetService', dodge=False)
axes[0].set_title('Global Customer Distribution')
axes[0].set_xlabel('')
axes[0].set_ylabel('Customer Count')
for c in axes[0].containers:
    axes[0].bar_label(c, fmt='%d', fontsize=9)

# % Churn wrt Global Churn
sns.barplot(data=df_t3, x='InternetService', y='%churn_wrt_Gc', ax=axes[1], hue='InternetService', dodge=False)
axes[1].set_title('Churn Distribution (wrt Global Churn)')
axes[1].set_xlabel('')
axes[1].set_ylabel('% of Total Churners')
for c in axes[1].containers:
    axes[1].bar_label(c, fmt='%.1f%%', fontsize=9)

# % Churn Typewise
sns.barplot(data=df_t3, x='InternetService', y='%churn_typewise', ax=axes[2], hue='InternetService', dodge=False)
axes[2].set_title('Churn Rate by Internet Type')
axes[2].set_xlabel('')
axes[2].set_ylabel('% within Internet Type')
for c in axes[2].containers:
    axes[2].bar_label(c, fmt='%.1f%%', fontsize=9)

for ax in axes:
    ax.tick_params(axis='x', rotation=0)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

#### Churn percentage by Gender

```python
df_g1 = df.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount')
df_gnc = df_t1.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount_nochurn')
df_gc = df_churnn.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount_churn')
df_g2 = df_g1.merge(df_gnc, on= 'gender', how= 'left').merge(df_gc, on= 'gender', how= 'left')
df_g2['%churn_genderwise'] = (df_g2['gendercount_churn']/df_g2['gendercount'])*100
df_total_churn = len(df_churnn)
df_g2['%churn_wrt_Gc'] = (df_g2['gendercount_churn']/df_total_churn)*100
df_g2
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14, 5))

colors = ['#6BAED6', '#FD8D3C'] 

# Total gender distribution
ax[0].pie(
    df_g2['gendercount'],
    labels=df_g2['gender'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor':'white', 'linewidth':1}
)
ax[0].set_title('Total Customers by Gender')

# No churn distribution
ax[1].pie(
    df_g2['gendercount_nochurn'],
    labels=df_g2['gender'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor':'white', 'linewidth':1}
)
ax[1].set_title('No Churn by Gender')

# Churned distribution
ax[2].pie(
    df_g2['gendercount_churn'],
    labels=df_g2['gender'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,
    wedgeprops={'edgecolor':'white', 'linewidth':1}
)
ax[2].set_title('Churned Customers by Gender')

plt.tight_layout()
plt.show()
```

#### Churn percentage by Contract

```python
df_ct1 = df.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount')
df_ctc = df_t1.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount_nochurn')
df_cc = df_churnn.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount_churn')
df_ct2 = df_ct1.merge(df_ctc, on= 'Contract', how= 'left').merge(df_cc, on= 'Contract', how= 'left')
df_ct2['%churn_Contractwise'] = (df_ct2['contractcount_churn']/df_ct2['contractcount'])*100
df_total_churn = len(df_churnn)
df_ct2['%churn_wrt_Gc'] = (df_ct2['contractcount_churn']/df_total_churn)*100
df_ct2
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Contract Type Distribution & Churn Insights', fontsize=14, fontweight='bold')

# Global distribution
sns.barplot(data=df_ct2, x='Contract', y='contractcount', ax=axes[0], hue='Contract', dodge=False)
axes[0].set_title('Global Customer Distribution')
axes[0].set_xlabel('')
axes[0].set_ylabel('Customer Count')
for c in axes[0].containers:
    axes[0].bar_label(c, fmt='%d', fontsize=9)

# % Churn wrt Global Churn
sns.barplot(data=df_ct2, x='Contract', y='%churn_wrt_Gc', ax=axes[1], hue='Contract', dodge=False)
axes[1].set_title('Churn Distribution (wrt Global Churn)')
axes[1].set_xlabel('')
axes[1].set_ylabel('% of Total Churners')
for c in axes[1].containers:
    axes[1].bar_label(c, fmt='%.1f%%', fontsize=9)

# % Churn Typewise
sns.barplot(data=df_ct2, x='Contract', y='%churn_Contractwise', ax=axes[2], hue='Contract', dodge=False)
axes[2].set_title('Churn Rate by Contract Type')
axes[2].set_xlabel('')
axes[2].set_ylabel('% within Contract Type')
for c in axes[2].containers:
    axes[2].bar_label(c, fmt='%.1f%%', fontsize=9)

for ax in axes:
    ax.tick_params(axis='x', rotation=0)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Internet Service Distribution & Churn Insights', fontsize=14, fontweight='bold')

# Global distribution
sns.barplot(data=df_p2, x='PaymentMethod', y='paymentmethodcount', ax=axes[0], hue='PaymentMethod', dodge=False)
axes[0].set_title('Global Customer Distribution')
axes[0].set_xlabel('')
axes[0].set_ylabel('Customer Count')
for c in axes[0].containers:
    axes[0].bar_label(c, fmt='%d', fontsize=9)

# % Churn wrt Global Churn
sns.barplot(data=df_p2, x='PaymentMethod', y='%churn_wrt_Gc', ax=axes[1], hue='PaymentMethod', dodge=False)
axes[1].set_title('Churn Distribution (wrt Global Churn)')
axes[1].set_xlabel('')
axes[1].set_ylabel('% of Total Churners')
for c in axes[1].containers:
    axes[1].bar_label(c, fmt='%.1f%%', fontsize=9)

# % Churn Typewise
sns.barplot(data=df_p2, x='PaymentMethod', y='%churn_PaymentMethodwise', ax=axes[2], hue='PaymentMethod', dodge=False)
axes[2].set_title('Churn Rate by Payment Method')
axes[2].set_xlabel('')
axes[2].set_ylabel('% within Payment Method')
for c in axes[2].containers:
    axes[2].bar_label(c, fmt='%.1f%%', fontsize=9)

for ax in axes:
    ax.tick_params(axis='x', rotation=25)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

#### Correlation

```python
df_encoded = df.copy()
df_encoded['churn_encoded'] = df['Churn'].replace({'Yes': 1, 'No': 0}).copy()
correlation = df_encoded[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'churn_encoded']].corr().round(2)
correlation
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5, fmt='.2f')
plt.title('Correlation Heatmap (Churn & Key Numeric Features)', fontsize=13, fontweight='bold')
plt.show()

# Correlation with Churn (Bar Chart)
plt.figure(figsize=(6,4))
correlation['churn_encoded'].drop('churn_encoded').sort_values(ascending=False).plot(
    kind='barh', color='teal')
plt.title('Feature Correlation with Churn', fontsize=13, fontweight='bold')
plt.xlabel('Correlation Coefficient')
plt.ylabel('')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
