#### Libraries Importation

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv(r"C:\Users\User\Documents\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df
```

#### Data Cleaning

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0).astype(float)
df
```

```python
df.describe()
```

#### Churn Percentage

```python
df_churn = df.groupby('Churn').size()
df_churn = df_churn.reset_index(name= 'count')
num = len(df)
df_churn['churn percentage'] = (df_churn['count']/num)*100
df_churn
```

```python
df_churnn = df[df['Churn'] == 'Yes']
df_churnn
```

```python
df_t1 = df[df['Churn'] == 'No']
df_t1
```

#### Churn percentage by Internet Service

```python
df_t2 = df.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount')
df_tnc = df_t1.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount_nochurn')
df_tc = df_churnn.groupby('InternetService').size().sort_values(ascending= False).reset_index(name= 'internetcount_churn')
df_t3 = df_t2.merge(df_tnc, on= 'InternetService', how= 'left').merge(df_tc, on= 'InternetService', how= 'left')
df_t3['%churn_typewise'] = (df_t3['internetcount_churn']/df_t3['internetcount'])*100
df_total_churn = len(df_churnn)
df_t3['%churn_wrt_Gc'] = (df_t3['internetcount_churn']/df_total_churn)*100
df_t3

```

#### Churn percentage by Gender

```python
df_g1 = df.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount')
df_gnc = df_t1.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount_nochurn')
df_gc = df_churnn.groupby('gender').size().sort_values(ascending= False).reset_index(name= 'gendercount_churn')
df_g2 = df_g1.merge(df_gnc, on= 'gender', how= 'left').merge(df_gc, on= 'gender', how= 'left')
df_g2['%churn_genderwise'] = (df_g2['gendercount_churn']/df_g2['gendercount'])*100
df_total_churn = len(df_churnn)
df_g2['%churn_wrt_Gc'] = (df_g2['gendercount_churn']/df_total_churn)*100
df_g2
```

#### Churn percentage by Contract

```python
df_ct1 = df.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount')
df_ctc = df_t1.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount_nochurn')
df_cc = df_churnn.groupby('Contract').size().sort_values(ascending= False).reset_index(name= 'contractcount_churn')
df_ct2 = df_ct1.merge(df_ctc, on= 'Contract', how= 'left').merge(df_cc, on= 'Contract', how= 'left')
df_ct2['%churn_Contractwise'] = (df_ct2['contractcount_churn']/df_ct2['contractcount'])*100
df_total_churn = len(df_churnn)
df_ct2['%churn_wrt_Gc'] = (df_ct2['contractcount_churn']/df_total_churn)*100
df_ct2
```

#### Churn percentage by Payment Method

```python
df_pm1 = df.groupby('PaymentMethod').size().sort_values(ascending= False).reset_index(name= 'paymentmethodcount')
df_pmnc = df_t1.groupby('PaymentMethod').size().sort_values(ascending= False).reset_index(name= 'paymentmethodcount_nochurn')
df_pc = df_churnn.groupby('PaymentMethod').size().sort_values(ascending= False).reset_index(name= 'paymentmethodcount_churn')
df_p2 = df_pm1.merge(df_pmnc, on= 'PaymentMethod', how= 'left').merge(df_pc, on= 'PaymentMethod', how= 'left')
df_p2['%churn_PaymentMethodwise'] = (df_p2['paymentmethodcount_churn']/df_p2['paymentmethodcount'])*100
df_total_churn = len(df_churnn)
df_p2['%churn_wrt_Gc'] = (df_p2['paymentmethodcount_churn']/df_total_churn)*100
df_p2
```
